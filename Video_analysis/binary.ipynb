```python
import cv2
import os
import numpy as np

# ì…ë ¥ í´ë” (í•œê¸€ ê²½ë¡œ í¬í•¨ ê°€ëŠ¥)
FOLDER_PATH = r"C:\Users\user\Downloads\ìƒˆ í´ë” (2)"

# ì¶œë ¥ í´ë”
OUTPUT_PATH = os.path.join(FOLDER_PATH, "binary")
os.makedirs(OUTPUT_PATH, exist_ok=True)

# ì„ê³„ê°’
THRESH = 100

# í•œê¸€ ê²½ë¡œ ì½ê¸° í•¨ìˆ˜
def hangul_imread(filename, flags=cv2.IMREAD_COLOR):
    try:
        n = np.fromfile(filename, np.uint8)
        img = cv2.imdecode(n, flags)
        return img
    except Exception as e:
        print(f"Error reading {filename}: {e}")
        return None

# í•œê¸€ ê²½ë¡œ ì €ì¥ í•¨ìˆ˜
def hangul_imwrite(filename, img, params=None):
    try:
        ext = os.path.splitext(filename)[1]
        result, n = cv2.imencode(ext, img, params)
        if result:
            with open(filename, mode='w+b') as f:
                n.tofile(f)
            return True
        return False
    except Exception as e:
        print(f"Error writing {filename}: {e}")
        return False

print(f"ğŸ“‚ Reading from: {FOLDER_PATH}")

count = 0
for file in os.listdir(FOLDER_PATH):
    if file.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".tif")):
        input_path = os.path.join(FOLDER_PATH, file)
        
        # [ìˆ˜ì •ë¨] í•œê¸€ ê²½ë¡œ ì½ê¸° í˜¸í™˜ í•¨ìˆ˜ ì‚¬ìš©
        img = hangul_imread(input_path, cv2.IMREAD_GRAYSCALE)

        if img is None:
            print(f"âŒ Failed to load: {file}")
            continue

        # ì´ì§„í™”
        _, binary = cv2.threshold(img, THRESH, 255, cv2.THRESH_BINARY)

        # ì €ì¥
        output_path = os.path.join(OUTPUT_PATH, file)
        
        # [ìˆ˜ì •ë¨] í•œê¸€ ê²½ë¡œ ì €ì¥ í˜¸í™˜ í•¨ìˆ˜ ì‚¬ìš©
        success = hangul_imwrite(output_path, binary)
        
        if success:
            print(f"âœ… Processed: {file}")
            count += 1
        else:
            print(f"âŒ Failed to save: {file}")

print(f"\nğŸ‰ Total {count} images converted successfully!")
```
