```python
import cv2
import os

# 입력 폴더
FOLDER_PATH = r"C:\Users\user\Downloads\compressed"

# 출력 폴더
OUTPUT_PATH = os.path.join(FOLDER_PATH, "binary")
os.makedirs(OUTPUT_PATH, exist_ok=True)

# 임계값 (아이콘 대비에 따라 140~180 조절 가능)
THRESH = 160

for file in os.listdir(FOLDER_PATH):
    if file.lower().endswith((".png", ".jpg", ".jpeg", ".bmp", ".tif")):
        input_path = os.path.join(FOLDER_PATH, file)

        # 흑백으로 읽기
        img = cv2.imread(input_path, cv2.IMREAD_GRAYSCALE)

        if img is None:
            continue

        # 이진화
        _, binary = cv2.threshold(
            img, THRESH, 255, cv2.THRESH_BINARY
        )

        # 저장
        output_path = os.path.join(OUTPUT_PATH, file)
        cv2.imwrite(output_path, binary)

print("✅ Binary conversion complete!")
```
