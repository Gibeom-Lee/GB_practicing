```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import cumulative_trapezoid
import os
from scipy.signal import find_peaks
```

```python
# 파일 경로
file_path = r"c:\Users\limye\OneDrive - gachon.ac.kr\HI Lab\0. Projects\0. On going\1. DC GENERATOR\0. 실험자료\5. Final stage\250314_저항\저항 별 csv\0ohm.csv"

# 데이터 로드 (헤더 없음)
df = pd.read_csv(file_path, header=None)

# NaN 값 제거 및 컬럼 이름 지정
df.iloc[:30, 1] = np.nan
df.columns = ["Time", "Voltage"]
df = df.astype(float)  # 데이터 타입 변환
```

```python
# 전압 데이터 추출
voltage = df['Voltage']

# RMS 전압 계산
rms_voltage = np.sqrt(np.mean(np.square(voltage)))
print(f"RMS Voltage: {rms_voltage:.4f} V")

# 그래프 그리기
plt.figure(figsize=(10, 6))
plt.plot(df['Time'], voltage, label='Original Voltage')
plt.hlines(rms_voltage, xmin=df['Time'].min(), xmax=df['Time'].max(),
           colors='r', linestyles='dashed', label=f'RMS Voltage ({rms_voltage:.4f} V)')
plt.title('AC Voltage and RMS Voltage')
plt.xlabel('Time (s)')
plt.ylabel('Voltage (V)')
plt.legend()
plt.grid(True)
plt.show()
```
