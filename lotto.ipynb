```python
import random
from collections import Counter

# List to store the results
results = []

# Repeating the random selection process 10,000 times
for _ in range(100000):
    random_numbers = random.sample(range(1, 46), 6)
    results.extend(random_numbers)

# Calculating the frequency of each number
counter = Counter(results)

# Selecting the 6 most common numbers
most_common_numbers = [num for num, count in counter.most_common(6)]
print(counter)
print( )
print(most_common_numbers)
```
