```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


```

```python
#FFT 실습

# 신호 생성 설정
Fs = 1000  # 샘플링 주파수 (Hz)
T = 1.0 / Fs  # 샘플링 간격
L = 1500  # 샘플 수
t = np.arange(0, L) * T  # 시간 벡터

print(t)
# 50 Hz와 120 Hz의 사인파 신호 생성
offset = 8
f1 = 50  # 주파수 1 (Hz)
f2 = 120  # 주파수 2 (Hz)
signal = offset + 3 * np.sin(2 * np.pi * f1 * t) + 3 * np.sin(2 * np.pi * f2 * t)
```

```python
# 노이즈 추가
noise = 2 * np.random.normal(size=t.shape)
signal_real = signal + noise

print(np.random.normal(size=t.shape))
# FFT 수행
fft_result = np.fft.fft(signal_real)
fft_freq = np.fft.fftfreq(L, T)
fft_amplitude = 2.0/L * np.abs(fft_result)

# 신호와 FFT 결과 시각화
plt.figure(figsize=(12, 8))

# 원 신호 및 노이즈 신호
plt.subplot(2, 1, 1)
plt.plot(t, signal_real, label='Noisy Signal')
plt.plot(t, signal, label='Original Signal', alpha=0.75)
plt.title('Time Domain Signal')
plt.xlabel('Time [s]')
plt.ylabel('Amplitude')
plt.legend()

# FFT 결과
plt.subplot(2, 1, 2)
plt.plot(fft_freq[:500//2], fft_amplitude[:500//2])
plt.title('Frequency Domain Signal')
plt.xlabel('Frequency [Hz]')
plt.ylabel('Amplitude')
plt.grid()

plt.tight_layout()
plt.show()
```
